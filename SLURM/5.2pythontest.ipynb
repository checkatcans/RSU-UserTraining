```python
# Cell 1: ตรวจสอบ GPU
import torch
print('PyTorch:', torch.__version__)
print('GPU available:', torch.cuda.is_available())
print('GPU name:', torch.cuda.get_device_name(0))
```

```python
# Cell 2: Matrix computation บน GPU
a = torch.rand(2000, 2000).cuda()
b = torch.rand(2000, 2000).cuda()
c = torch.matmul(a, b)
print('Result shape:', c.shape)
print('Device:', c.device)
```

```python
# Cell 3: Monitor GPU memory
print('GPU Memory Allocated:', torch.cuda.memory_allocated(0)/1024**2, 'MB')
print('GPU Memory Reserved:', torch.cuda.memory_reserved(0)/1024**2, 'MB')
```
